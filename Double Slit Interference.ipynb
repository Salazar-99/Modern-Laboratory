```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
laser_data = np.loadtxt("data/YDS.txt")
bulb_data = np.loadtxt("data/OPAAT.txt")
bulb_noise = np.loadtxt("data/OPAAT-Noise.txt")
```

```python
#Plotting laser data
laser_data_x = laser_data[:,0]
laser_data_y = laser_data[:,1]
laser_data_plot = plt.plot(laser_data_x, laser_data_y)
plt.xlabel("Slit position (mm)", fontweight = "bold")
plt.ylabel("Voltage (V)", fontweight = "bold")
```

```python
#Subtracting noise measurement from bulb data
average_noise = np.mean(bulb_noise)
bulb_data_x = bulb_data[:,0]
bulb_data_y = bulb_data[:,1]
calibrated_bulb_data_y = bulb_data_y - average_noise
print(average_noise)
```

```python
#Plotting calibrated bulb data
bulb_data_plot = plt.plot(bulb_data_x, calibrated_bulb_data_y)
plt.xlabel("Slit position (mm)", fontweight = "bold")
plt.ylabel("Photon Counts", fontweight = "bold")
```

It seems that after about 6mm the data is skewed due to the slit cover revealing the photomultiplier so I will cut it out and replot. The data was taken in .25mm increments so first I will calculate the correct index.

```python
#Cutting out extraneous data
slice_index = 24
reshaped_bulb_data_x = bulb_data_x[:slice_index] 
reshaped_bulb_data_y = calibrated_bulb_data_y[:slice_index]
```

```python
#Plotting reshaped data
reshaped_bulb_data_plot = plt.plot(reshaped_bulb_data_x, reshaped_bulb_data_y)
plt.xlabel("Slit position (mm)", fontweight = "bold")
plt.ylabel("Photon Counts", fontweight = "bold")
```
